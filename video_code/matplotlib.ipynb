```python
import matplotlib.pyplot as plt
import numpy

x = numpy.linspace(-5, 5, 100)

f = numpy.sin(x)

f
```

```python
import numpy
import matplotlib.pyplot as plt

plt.figure()
plt.plot(x, f)
plt.title("Sin(x)")
plt.xlabel("x")
plt.ylabel("f")
```

```python
import numpy
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes()
ax.plot(x, f)
ax.set(xlim=(-10, 10), ylim=(-5, 5), xlabel="x", ylabel="y")

# More info at "anatomy of a figure" on the web
```
