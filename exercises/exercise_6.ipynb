```python
# 0. Unit conversion

class Conversion:
    def __init__(self, value):
        self.value = value
    
    @property
    def value(self):
        return self._value
    
    @value.setter
    def value(self, new_value):
        if not isinstance(new_value, (int, float)):
            raise ValueError("Only numbers")
        self._value = new_value
    
    def inch_to_cm(self):
        return f"{self.value} inches in centimeters is {self.value * 2.54}"
    
    def foot_to_meters(self):
        return f"{self.value} feet in meters is {self.value * 0.3048}"
    
    def pound_to_kg(self):
        return f"{self.value} pounds in kilograms is {self.value * 0.45359237}"
    
    def __repr__(self):
        return f"Value = {self.value}"


try:
    a = Conversion(10)
except ValueError as err:
    print(err)

a.pound_to_kg()
```

```python
# 1. Person

import re

class Person:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email
    
    @property
    def name(self):
        return self._name
    @name.setter
    def name(self, new_name):
        if not isinstance(new_name, str):
            raise ValueError("Name can only be string")
        self._name = new_name
    
    @property
    def age(self):
        return self._age
    @age.setter
    def age(self, new_age):
        if not isinstance(new_age, (float, int)):
            raise ValueError("Age must be a number")
        else:
            if new_age < 0 or new_age > 125:
                raise ValueError("Age must be between 0 and 125")
        self._age = new_age
    
    @property
    def email(self):
        return self._email
    @email.setter
    def email(self, new_email):
        if not isinstance(new_email, str):
            raise ValueError("Email must be a string")
        else:
            if bool(re.search("@", new_email)) == False:
                raise ValueError("Email must contain a '@'")
        self._email = new_email
    
    def __repr__(self):
        return f"Name: {self.name}\nAge: {self.age}\nEmail: {self.email}"
    
    def say_hello(self):
        return f"Hi, my name is {self.name}, i am {self.age} years old and my email address is {self.email}"

try:
    a = Person("Karl", 23, "kalle.rydberg02@gmail.com")
except ValueError as err:
    print(err)

a.say_hello()
```
