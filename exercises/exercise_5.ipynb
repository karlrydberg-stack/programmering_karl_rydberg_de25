```python
# 0. Area

def triangle_area(a, b):
    return a*b/2

print(triangle_area(2, 2))
```

```python
# 1. Euclidean distance
import math

def euclidean():
    while True:
        try:
            x1 = float(input("X-coordinate nr. 1"))
            y1 = float(input("Y-coordinate nr. 1"))
            x2 = float(input("X-coordinate nr. 2"))
            y2 = float(input("Y-coordinate nr. 2"))
            break
        except ValueError:
            print("Only numerical input")
    
    return math.sqrt((x1-y1)**2 - (x2-y2)**2)

print(euclidean())
```

```python
# 2. Mathematical function
import matplotlib.pyplot as plt
import numpy

def f(x):
    return x ** 2 - 3

def g(x):
    return 4 * x - 7

x1 = numpy.array([f(x) for x in range(-10, 11)])
x2 = numpy.array([g(x) for x in range(-10, 11)])


plt.plot(x1)
plt.plot(x2)

plt.show() 
```
